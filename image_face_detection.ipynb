```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
photoo = cv2.imread('thor.jpg')
```

```python
plt.imshow(photoo)
```

```python
rgbwala = cv2.cvtColor(photoo, cv2.COLOR_BGR2RGB)
plt.imshow(rgbwala)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
faces = classifier.detectMultiScale(photoo)
```

```python
face = faces[0]
```

```python
x, y, w, h = face
face = rgbwala[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```
